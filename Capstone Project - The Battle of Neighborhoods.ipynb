# Capstone Project - The Battle of Neighborhoods

## Introduction/Business Problem

Toronto, the capital of the province of Ontario, is a major Canadian city along Lake Ontario’s northwestern shore. New York City (NYC), on the other hand, is the most populous city in the United States. New York and Toronto are similar in many ways. Both cities are financial capitals of their countries and throughout the years, their neighborhoods have evolved and transformed, offering a wide variety of venues to their inhabitants. Toronto and New York are both incredibly diverse with huge immigrant populations and are the most crowded cities in their respective countries. They also both expensive and dynamic. However, are these cities actually so similar?

The purpose of this whole exercise is for submission of the final capstone project for the "IBM Data Science" course on Coursera as well as to determine if Toronto and New York are as similar as everyone says they are.
